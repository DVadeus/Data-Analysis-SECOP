### Contratos duplicados

```python
import duckdb
```

Extracción de los IDs duplicados en toda la base de datos

```python
contratos_duplicados = duckdb.query("""
    SELECT id_contrato
    FROM '../data/raw/**/*.parquet'
    WHERE id_contrato IN(
    SELECT id_contrato
    FROM '../data/raw/**/*.parquet'
    GROUP BY id_contrato
    HAVING COUNT(*) > 1
    )
""").df()['id_contrato']
```

```python
contratos_duplicados_series = duckdb.query("""
    SELECT id_contrato
    FROM '../data/raw/**/*.parquet'
    WHERE id_contrato IN(
    SELECT id_contrato
    FROM '../data/raw/**/*.parquet'
    GROUP BY id_contrato
    HAVING COUNT(*) > 1
    )
""").df()
```

```python
contratos_duplicados_series
```

```python
for contrato in contratos_duplicados_series.itertuples():
    print(contrato.id_contrato)
```

```python
df_list = []

for contrato in contratos_duplicados_series.itertuples():
    df_list.append(pd.DataFrame(duckdb.query(f"""
    SELECT *
    FROM '../data/raw/**/*.parquet'
    WHERE id_contrato = '{contrato.id_contrato}'
    """).df()))
    print(f"Contrato {contrato} añadido a la lista")
contratos_duplicados = pd.concat(df_list, ignore_index=True)
```

```python
contratos_duplicados.to_excel("contratos_duplicados.xlsx", index= False)
```
